```python
import os
import json
```

```python
labels = os.listdir("./labels_old/")
```

```python
for each in labels:
    with open("./labels_old/"+each, "r", encoding="utf-8") as f:
        content = f.read()
        label = json.loads(content)
        imgName = each.split(".")[0]
        print(label['imagePath'])
        print(imgName)
        label['imagePath'] = "../pictures/"+imgName+".TIF"
        print("done: imagePath=", label['imagePath'])
        
        f2 = open("./labels/"+each, 'w', encoding="utf-8")
        f2.write(json.dumps(label))
    f2.close()
    f.close()
```

```python
for each in labels:
    with open("./labels/"+each, 'w') as f:
        label = json.load(f)
        path = label['imagePath']
        imgName = path.split('/')[-1]
        label['imagePath'] = "../pictures/"+imgName

        json.dump("./labels_new/"+each, label)
```
